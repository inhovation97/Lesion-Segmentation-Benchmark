# module import

```python
import glob
import matplotlib.pyplot as plt
import matplotlib.image as img
from PIL import Image
import cv2
import matplotlib.gridspec as gridspec
import random
```

# data path

```python
# image
wound_path = sorted(glob.glob('/data/segmentation/wound/trainset/images/*'))
ISIC_path = sorted(glob.glob("/data/segmentation/ISIC-2017/trainset/images/*"))
CVC_path = sorted(glob.glob('/data/segmentation/CVC-ClinicDB/trainset/images/*'))
Kvasir_path = sorted(glob.glob("/data/segmentation/Kvasir-SEG/trainset/images/*"))
benign_path = sorted(glob.glob("/data/segmentation/breast-cancer/trainset_benign/images/*"))
malignant_path = sorted(glob.glob("/data/segmentation/breast-cancer/trainset_malignant/images/*"))

# mask
wound_M_path = sorted(glob.glob('/data/segmentation/wound/trainset/labels/*'))
ISIC_M_path = sorted(glob.glob("/data/segmentation/ISIC-2017/trainset/labels/*"))
CVC_M_path = sorted(glob.glob('/data/segmentation/CVC-ClinicDB/trainset/labels/*'))
Kvasir_M_path = sorted(glob.glob("/data/segmentation/Kvasir-SEG/trainset/labels/*"))
benign_M_path = sorted(glob.glob("/data/segmentation/breast-cancer/trainset_benign/labels/*"))
malignant_M_path = sorted(glob.glob("/data/segmentation/breast-cancer/trainset_malignant/labels/*"))
```

# data load

```python
i = 5
# image
wound = cv2.imread(wound_path[6])  # 2.jpg
wound = cv2.cvtColor(wound, cv2.COLOR_BGR2RGB)

ISIC = cv2.imread(ISIC_path[5])
ISIC = cv2.cvtColor(ISIC, cv2.COLOR_BGR2RGB)

CVC = cv2.imread(CVC_path[i])
CVC = cv2.cvtColor(CVC, cv2.COLOR_BGR2RGB)

Kvasir = cv2.imread(Kvasir_path[i+2])
Kvasir = cv2.cvtColor(Kvasir, cv2.COLOR_BGR2RGB)

benign = cv2.imread(benign_path[i])
benign = cv2.cvtColor(benign, cv2.COLOR_BGR2RGB)

malignant = cv2.imread(malignant_path[i])
malignant = cv2.cvtColor(malignant, cv2.COLOR_BGR2RGB)


# mask
wound_M = cv2.imread(wound_M_path[6])


ISIC_M = cv2.imread(ISIC_M_path[2])


CVC_M = cv2.imread(CVC_M_path[i])


Kvasir_M = cv2.imread(Kvasir_M_path[i+2])


benign_M = cv2.imread(benign_M_path[i])


malignant_M = cv2.imread(malignant_M_path[i])
```

```python
fig, axs = plt.subplots(2,3)


axs[0,0].imshow(ISIC)
axs[0,0].set_xticks([])
axs[0,0].set_yticks([])




axs[1,0].imshow(wound)
axs[1,0].set_xticks([])
axs[1,0].set_yticks([])



axs[0,1].imshow(CVC)
axs[0,1].set_xticks([])
axs[0,1].set_yticks([])



axs[1,1].imshow(Kvasir)
axs[1,1].set_xticks([])
axs[1,1].set_yticks([])


axs[0,2].imshow(benign)
axs[0,2].set_xticks([])
axs[0,2].set_yticks([])



axs[1,2].imshow(malignant)
axs[1,2].set_xticks([])
axs[1,2].set_yticks([])

fig.tight_layout()
plt.show()


```
