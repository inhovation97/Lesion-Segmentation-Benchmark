# ISIC-2017 : 피부암 데이터

```python
from ploting import *
import random

import glob
import os

train_images = glob.glob('/data/segmentation/ISIC-2017/trainset/images/*')
train_labels = glob.glob('/data/segmentation/ISIC-2017/trainset/labels/*')
train_images = [img for img in train_images if img.find('jpg')!= -1] # super pixels 이미지 제외

valid_images = glob.glob('/data/segmentation/ISIC-2017/validationset/images/*')
valid_labels = glob.glob('/data/segmentation/ISIC-2017/validationset/labels/*')
valid_images = [img for img in valid_images if img.find('jpg')!= -1] # super pixels 이미지 제외

test_images = glob.glob('/data/segmentation/ISIC-2017/testset/images/*')
test_labels = glob.glob('/data/segmentation/ISIC-2017/testset/labels/*')
test_images = [img for img in valid_images if img.find('jpg')!= -1] # super pixels 이미지 제외

i = random.randint(0, len(valid_labels))

valid_img_path = sorted(valid_images)[i]
valid_label_path = sorted(valid_labels)[i]

print('trainset 이미지 개수:{}\nvalidset 이미지 개수:{}, \ntestset 이미지 개수:{}'.format(len(train_images), 
                                                                           len(valid_images),
                                                                           len(test_images)))
masked_raw_imgplot(valid_img_path, valid_label_path)
```
